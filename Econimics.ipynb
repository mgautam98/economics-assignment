```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
excel_file = "./data.XLSX"
```

```python
data = pd.read_excel(excel_file)
```

```python
data.head()
```

```python
data = data.drop(index=38)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(32, 16), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data['year'], data['value'])
plt.ylabel("In Billion")
plt.xlabel("Year")
plt.show()
```

```python
# Growth rate
rates = []
rates.append(0)
for i in range(1,38):
    rates.append(data['value'][i] - data['value'][i-1])
```

```python
figure(num=None, figsize=(32, 16), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data['year'], rates)
plt.ylabel("Rate in Billion/year")
plt.xlabel("Year")
plt.show()
```
